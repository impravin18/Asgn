```python
#importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Loading dataset into pandas dataframe
df1=pd.read_csv('covid_19_india.csv')
```

```python
#Displays first 5 rows 
df1.head()
```

```python
#Displays last 5 rows
df1.tail()
```

```python
#Returns the number of null values in each column
df1.isnull().sum()
```

```python
#Used to get the initial statistics
df1.describe()
```

```python
#Returs a tuple conataining dimensios i.e no of rows and no of columns
df1.shape
```

```python
df1.groupby('Date').count()
```

```python
#Used to get mean of cured and deaths grouping by State/Union Territory
df1.groupby(df1['State/UnionTerritory'])[['Cured','Deaths']].mean()
```

```python
#Max value group by State/UnionTerritory
df1.groupby(df1['State/UnionTerritory'])[['Cured','Deaths']].max()
```

```python
#Min value group by State/UnionTerritory
df1.groupby(df1['State/UnionTerritory'])[['Cured','Deaths']].min()
```

```python
#Standard Deviation group by State/UnionTerritory
df1.groupby(df1['State/UnionTerritory'])[['Cured','Deaths']].std()
```

```python
#Median value group by State/UnionTerritory
df1.groupby(df1['State/UnionTerritory'])[['Cured','Deaths']].median()
```

```python
#Inital Statistics group by State/UnionTerritory
df1.groupby(df1['State/UnionTerritory'])['Cured'].describe()
```

# Data Visualization

```python
#Scatter Plot
x1 = df1['State/UnionTerritory']
y1 = df1['Deaths']
```

```python
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(14)
plt.scatter(y1, x1, marker='+',label="Deaths")
plt.xlabel("Deaths")
plt.ylabel("State/UnionTerritory")
plt.legend()
plt.show()
```

```python
#Bar Graph
x1 = df1['State/UnionTerritory']
y1 = df1['Cured']
```

```python
f = plt.figure()
f.set_figwidth(50)
f.set_figheight(14)
plt.bar(x1, y1,label="Cured")
plt.xlabel("Cured")
plt.ylabel("State/UnionTerritory")
plt.legend()
plt.show()
```

```python
#Displays last 5 rows
df2.tail()
```

```python
#Mean value group by gender
df2.groupby('gender')[['math score','reading score','writing score']].mean()
```

```python
#Max value group by gender
df2.groupby('gender')[['math score','reading score','writing score']].max()
```

```python
#Min value group by gender
df2.groupby('gender')[['math score','reading score','writing score']].min()
```

```python
#Standard Deviation group by gender
df2.groupby('gender')[['math score','reading score','writing score']].std()
```

```python
#Median value group by gender
df2.groupby('gender')[['math score','reading score','writing score']].median()
```

```python
#Quantile value group by gender
df2.groupby('gender')[['math score','reading score','writing score']].quantile()
```

```python
#Mode value group by gender
df2.groupby('gender').agg(lambda x:x.value_counts().index[0])
```

```python
x1 = df2['gender']
y1 = df2['math score']
```

```python
plt.scatter(y1, x1, marker='+',label="Math Score")
plt.ylabel("Gender")
plt.xlabel("Scores")
plt.legend()
plt.show()
```

```python
x1 = df2['gender']
y1 = df2['reading score']
plt.scatter(y1, x1, marker='+',label="Reading Score",color='r')
plt.ylabel("Gender")
plt.xlabel("Scores")
plt.legend()
plt.show()
```

```python
x1 = df2['gender']
y1 = df2['writing score']
plt.scatter(y1, x1, marker='+',label="Writing Score",color='orange')
plt.ylabel("Gender")
plt.xlabel("Scores")
plt.legend()
plt.show()
```
